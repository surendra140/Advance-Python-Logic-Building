**Factorial of a number**

```python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n-1)
    #          5   factorial(4)
factorial(5)
```

**Fibonacii Series**

```python
def fibonacii(n):
    if n == 0 or n == 1:
        return n
    else:
        return fibonacii(n-1) + fibonacii(n-2)

fibonacii(5)
```

# Dynamic Programming

memoization

```python
d = {'key': 10}
```

```python
def fibonacii(n, memo = {}):
    if n == 0 or n == 1:
        return n
    
    if n > 1:
        if n in memo:
            return memo[n]
        else:
            result = fibonacii(n-1, memo) + fibonacii(n-2, memo)
            memo[n] = result
            return result
```

```python
fibonacii(120)
```

# Functions are Objects

```python
def shout(word = 'Hello'):
    return word + '!!'

print(shout())

scream = shout

print(scream())

del shout

try:
    shout()
except NameError as e:
    print(e)
    
print(scream())
```

```python
def advance():
    
    def cat(word = 'hi'):
        return word + '##'
    
    return cat()

print(advance())

try:
    print(cat())
except NameError as e:
    print(e)
```

```python
def getFunction(other = 'shout'):
    
    def shout(word = 'Hello'):
        return word + '!!'
    
    def cat(word='Hi'):
        return word + '##'
    
    if other == 'shout':
        return shout
    else:
        return cat

t = getFunction()
print(t())
print(t)
```

```python
class A:
    def add(self):
        print("Add A")
    
class B:
    def add(self):
        print("Add B")
    
class C(A,B):
    def show(self):
        print("show C")
    
z = C()

z.show()
z.add()
```

```python
class A:
    def add(self):
        print("Add A")
    
class B:
    def add(self):
        print("Add B")
    
class C(B,A):
    def show(self):
        print("show C")
    
z = C()

z.show()
z.add()
```
